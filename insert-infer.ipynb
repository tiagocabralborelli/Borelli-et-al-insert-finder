```python
from Bio import SeqIO
import glob
import os
import pandas as pd
import matplotlib.pyplot as plt
from Bio.Blast.Applications import NcbiblastnCommandline
import customfunc
```

```python
class primer_list ():
    
    
    def __init__(self,diretorio):
        import glob
        import os
        from Bio import SeqIO
        self.diretorio = diretorio
        abi = glob.glob(diretorio + "/*.ab1")
        self.abi = abi
        if len(abi) == 0:
            print('Lista de abi vazia. Escolha outro diretorio')
            
    def to_fasta(self,NomePasta):
        import glob
        import os
        from Bio import SeqIO
        os.system(f'mkdir {self.diretorio}/{NomePasta}')
        
        for arquivos_abi in self.abi:
            f = SeqIO.convert(f'{arquivos_abi}', 'abi', f'{arquivos_abi}.fasta','fasta')
            os.system(f'mv {arquivos_abi}.fasta {NomePasta}')
        fastas = glob.glob(f"{self.diretorio}/{NomePasta}" + '/*.fasta')
        
        for arquivos in fastas:
            for nomes in SeqIO.parse(f'{arquivos}','fasta'):
                nome = arquivos.split('/')[-1].split('.')[0] + '.' + arquivos.split('/')[-1].split('.')[1]
                novo = open(f'{nome}.fasta','a')
                ident = nome.split('.')[0]
                novo.write(f'>{ident}\n')
                novo.write(f'{str(nomes.seq)}')
                novo.close()
                os.system(f'rm {arquivos}')
                os.system(f'mv {nome}.fasta {self.diretorio}/{NomePasta}')
                
        fastas = glob.glob(f"{self.diretorio}/{NomePasta}" + '/*.fasta')
        
        if len(fastas) == 0:
            print('pasta com fastas vazia')
        else:
            self.fastas = fastas
            
        forward = [f for f in self.fastas if "F" in f.split('/')[-1].split('.')[0]]
        self.forward = forward
        forward.sort()
        
        reverse = [r for r in self.fastas if "R" in r.split('/')[-1].split('.')[0]]
        self.reverse = reverse
        reverse.sort()
        
        if len(forward) != len(forward):
            print('As listas forward e reverse tem tamanhos diferentes')            
```

```python
class blast_sequences ():
    
    def __init__(self,f,r,diretorio):
        import os
        self.r = r
        self.f = f
        self.diretorio = diretorio
        
        
    def blast(self,modo,genoma,*lista):
        from Bio.Blast.Applications import NcbiblastnCommandline
        import os
        self.lista = lista
        self.modo = str(modo).upper()
        self.genoma = genoma
        os.system(f'mkdir {self.diretorio}')
        
        if modo == '_F':
            for primers in self.f:
                for identificadores in self.lista:
                    if primers.split('/')[-1].split('.')[0].endswith(identificadores):
                        c = NcbiblastnCommandline(query = f'{primers}', subject = genoma ,out = f"{primers}_{genoma}.csv", outfmt = 6)
                        c()
                        os.system(f'mv {primers}_{genoma}.csv {self.diretorio}')
                   
                    
        elif modo == '_R':
            for primers in self.r:
                for identificadores in self.lista:
                    if primers.split('/')[-1].split('.')[0].endswith(identificadores):
                        c = NcbiblastnCommandline(query = f'{primers}', subject = genoma ,out = f"{primers}_{genoma}.csv", outfmt = 6)
                        c()
                        os.system(f'mv {primers}_{genoma}.csv {self.diretorio}')
        
```

```python
arquivos =  glob.glob("/home/tiagoboreli/Documentos/trimopython/blasts" + '/*.csv')
lista_df = []
for arqs in arquivos:
    df = pd.read_csv(arqs, index_col=None, header=None, sep = '\t',nrows = 1 ,names = ['query acc.ver', 'subject acc.ver','identity', 'alignment length', 'mismatches',' gap opens', 'q. start', 'q. end', 's. start', 's. end', 'evalue', 'bit score'])
    df['arquivo'] = arqs
    lista_df.append(df)
genomes = pd.concat(lista_df,axis=0, ignore_index=True)
```

```python
genomes_f = genomes.set_index('query acc.ver').filter(like = '_F', axis = 0)
```

```python
genomes_f.reset_index(inplace = True)
```

```python
genomes_r = genomes.set_index('query acc.ver').filter(like = '_R', axis = 0)
```

```python
strac = pd.DataFrame({'Primer_f':genomes_f['query acc.ver'],
                      'Primer_r':genomes_r['query acc.ver'],
                      'Nodes':genomes_f['subject acc.ver'] + '/' + genomes_r['subject acc.ver'],
                     'Genoma':genomes_r['arquivo'].apply(lambda x: x.split('/')[-1].split('.')[2].split('_')[1])})
```

```python
l = 0
i = 0
start = []
End = []
while l < len(genomes_r):
    p = [genomes_f['s. start'][i], genomes_f['s. end'][i],genomes_r['s. start'][i], genomes_r['s. end'][i]]
    start.append(min(p))
    End.append(max(p))
    p = []
    l += 1
    i += 1
    
    
    
```

```python
strac['Start'] = start
strac['End'] = End
strac['Length'] = strac['End'] - strac['Start']
```

```python
validation = []
for items in strac.itertuples():
    if items.Nodes.split('/')[0] == items.Nodes.split('/')[1]:
        validation.append('True')
    else:
        validation.append('False')
strac['Validation'] = validation
```
